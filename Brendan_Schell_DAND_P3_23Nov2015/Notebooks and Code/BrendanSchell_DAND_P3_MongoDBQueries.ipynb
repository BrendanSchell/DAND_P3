```python
# connect to MongoDB

from pymongo import MongoClient

client = MongoClient('localhost:26016')



```

```python
# set database and collection to use

db = client['street_data']

coll = db.P3
```

```python
#use of street field

cursor = coll.aggregate(
    [
        {'$group': {'_id': '$address.street', 'count' : {'$sum' : 1}}},
        {'$sort' : {'count' : -1}},
        {'$limit' : 5}
    ]
)
for doc in cursor:
    print doc
```

```python
coll.find({'address.city': 'City of Toronto'})
```

```python
#entries with postal codes

cursor =  coll.aggregate(
    [
        {'$match' : {'postcode':{'$exists' : 1}}},
        {'$limit' : 100}
    ])
for i in cursor:
         print i
```

```python
# top level fields

reduce(
    lambda all_keys, rec_keys: all_keys | set(rec_keys), 
    map(lambda d: d.keys(), coll.find()), 
    set()
)
```

```python
for i in coll.find({'cuisine:coffee' : {'$exists' : 1}}):
    print i
```

```python
cursor = coll.aggregate(
    [
        {'$group': {'_id': '$diet:gluten_free', 'count' : {'$sum' : 1}}},
        {'$sort' : {'count' : -1}},
        {'$limit' : 5}
    ]) 
for i in cursor:
    print i
```
